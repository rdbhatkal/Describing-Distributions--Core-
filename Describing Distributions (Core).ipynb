# Load the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Loading the dataset
url = "Data/data.csv"
df = pd.read_csv(url)
```

```python
# Looking at first 5 rows of the dataframe
df.head()
```

```python
# Looking at the dataset
df.info()
```

Create a function to plot a histogram

```python
def plot_graph(col):
  feature = df[col]
  mean = feature.mean()
  median = feature.median()
  std = feature.std()
  plus_one_std = mean + std
  minus_one_std = mean - std
  fig,ax = plt.subplots(figsize=(12,10))
  sns.histplot(feature ,ax=ax,stat='density', kde = True)
  ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
  ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
  ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
  ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
  ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
  ax.set_title(f'Density distribution for {col}')
  ax.legend();
```

## Plot VitD_levels

### Plot a histogram with a kde (kernel density estimate)


```python
# Call the function to plot VitD_levels
plot_graph('VitD_levels')
```

### Answer the following questions

* Is it Discrete or Continuous?

    **It is a continuous distribution**




* Does it have a skew? If so, which direction (+/-)

    **It does not have a skew. The mean and median are almost similar**


```python
# calculate the kurtosis 
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt
```

* What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

    **Given our Kurtosis value is ~ 3, it is a Mesokurtic Kurtosis**
